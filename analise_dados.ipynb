### **Aula 4.1: Estatística Básica**

```python
#Biblioteca Numpy: permite fazer análises estatísticas e forma rápida e fácil

import numpy as np
```

```python
lista_idades = [26, 30, 32, 22, 26, 35, 40, 20, 43, 31, 23]
```

```python
#Média: A média é um valor que representa o "meio" de um conjunto de números. É usada para entender o comportamento geral ou o padrão de um grupo de dados.
'''Como calcular:
Some todos os números.
Divida o total pelo número de elementos no grupo.'''

np.sum(lista_idades) #Estamos realizando a soma das idades

len(lista_idades) #Estamos contando a quantidade de idades

media = np.sum(lista_idades)/len(lista_idades) #Cálculo da média das idades
media
```

```python
#outra forma e fazer a média, podemos usar mean

media = np.mean(lista_idades)
print('O valor da média das idades é:', media)
```

```python
'''
A **mediana** é o número que fica bem no meio de um conjunto de valores quando eles estão organizados em ordem crescente ou decrescente. Ela divide os dados em duas partes iguais: metade dos valores fica abaixo da mediana e metade fica acima.

### Como encontrar a mediana:
1. Organize os números em ordem crescente.
2. Se o número total de valores for:
   - **Ímpar**: A mediana é o valor que está exatamente no meio.
   - **Par**: A mediana é a média dos dois valores centrais.

### Exemplo:
- Conjunto ímpar: \( 3, 7, 9 \) → A mediana é \( 7 \).
- Conjunto par: \( 2, 4, 6, 8 \) → A mediana é \( (4 + 6) \div 2 = 5 \).

A mediana é útil para evitar que valores extremos distorçam os resultados, como acontece na média.


'''

#ordenar a lista
np.sort(lista_idades)

mediana = np.median(lista_idades)
print('O valor da mediana das idades é:', mediana)
```

```python
#Outro exemplo mudando a lista inserida:
lista_idades = [26, 30, 32, 22, 26, 35, 40, 20, 43, 31, 23, 100]
lista_idades.sort()

mediana = np.median(lista_idades)
print('O valor da mediana das idades é:', mediana)
```

### **Aula 4.3: Tratando valores discrepantes - outliers (parte 1)**

##### **Média, Desvio Padrão, limite inferior e limite superior**

```python
#Entendendo o conceito

lista_idades = [26,30,32,22,26,35,40,20,43,31,23]
np.mean(lista_idades)
```

```python
#Vamos supor que a pessoa ao digitar a lista, escreveu 400 no lugar de 40. Vamos observar que o valor da média obtido é bem maior do que o valor correto. O 400 é considerao
#um outlier

lista_idades = [26,30,32,22,26,35,400,20,43,31,23]
media = np.mean(lista_idades)
media
```

```python
'''
Temos algumas formas de identificar esses valores, nessa aula aprendemos duas:
➡️Utilizando o desvio padrão - métodos estatísticos
➡️Através da regra dos quartis
Devemos usá-las de acordo com o contexto dos nossos dados, se eles já são bem variados ou não.

'''

#Vamos fazer o desvio-padrão (std) com a lista contendo o valor 400. Podemos Notar que o valor do desvio é muito alto para os valores que temos. Dessa forma, podemos concluir que temos um outlier entre os dados.
#Pela regra do desvio padrão, quanto mais próximo de zero, mais confiáveis são os dados.
desvio = np.std(lista_idades )
desvio
```

```python
#Para confirmar as hipóteses, vamos usar a média e o desvio padrão para determinar os limites inferior e superior.

limite_inferior = media - (3 * desvio)
limite_inferior
```

```python
limite_superior = media + (3 * desvio)
limite_superior
```

```python
#Dessa forma, todos os valores acima desses extremos são considerados outliers. Como 400 está acima do limite superior ele é um outlier.
```

##### **Quartis e Bloxpot**

```python
'''
Quartis são medidas estatísticas que dividem um conjunto de dados ordenados em quatro partes iguais, com cada parte representando 25% da distribuição dos valores.
Eles são utilizados para entender a dispersão, a centralidade e a distribuição de um conjunto de dados.
Os quartis são definidos como:
1. Primeiro Quartil (Q1 ou 25º Percentil): O valor que separa os 25% menores dos 75% maiores valores do conjunto de dados. Ele é o ponto abaixo do qual estão os primeiros 25% dos dados.
2. Segundo Quartil (Q2 ou Mediana ou 50º Percentil): O valor central do conjunto de dados quando ordenado. Ele divide o conjunto em duas metades iguais, com 50% dos dados abaixo e 50% acima.
3. Terceiro Quartil (Q3 ou 75º Percentil): O valor que separa os 75% menores dos 25% maiores valores do conjunto. Ele é o ponto abaixo do qual estão os primeiros 75% dos dados.
Como calcular os quartis:
Organize os dados em ordem crescente.
Identifique a posição de cada quartil:

Q1: Divida o conjunto de dados em duas metades, e o ponto médio da metade inferior é 𝑄1
Q2: É a mediana geral do conjunto de dados.
Q3: Divida o conjunto em duas metades, e o ponto médio da metade superior é Q3.

Para visualizar utilizamos o Bloxpot.
'''

#Importação da biblioteca matplotlib

import matplotlib.pyplot as plt
```

```python
#Plotando o bloxpot para visualização dos quartis

plt.boxplot(lista_idades)

#Linha amarela: mediana
#Bordas das caixinhas: Quartis Q1 e Q3. As linhas são os limites inferior e superior. O ponto lá em cima é o outlier.
```

### **4.4 - Intervalo de confiança e distribuição amostral**

```python
'''
➡️ Intervalo de confiança: É um intervalo de valores, calculado a partir de dados amostrais, que tem uma probabilidade específica (geralmente 95% ou 99%)
de conter o valor verdadeiro de um parâmetro populacional, como a média ou proporção. Ele indica a margem de erro associada à estimativa.
Importância do Intervalo de Confiança:
Precisão: O intervalo de confiança nos dá uma faixa dentro da qual acreditamos que a média populacional se encontra, tornando nossa estimativa mais informativa do que uma única média amostral.
Incerteza: Reconhece a incerteza inerente à amostragem e nos dá uma ideia de quanto essa incerteza pode ser.

➡️ Distribuição amostral: É a distribuição de uma estatística (como média ou proporção) calculada a partir de todas as possíveis amostras de um mesmo tamanho
extraídas de uma população. Ela mostra como essa estatística varia de amostra para amostra e é fundamental para calcular intervalos de confiança e realizar testes de hipóteses.
'''
```

```python
#Pegando todos os dados da nossa tabela e colocando em uma variável chamada de salário

salarios = dataframe['SALARIO']
salarios
```

### **4.6 - Correlação, diferentes funções para dados discretos e contínuos**

```python
'''
A correlação nos diz se e como duas variáveis estão relacionadas🤔 Existem dois tipos principais de correlação: positiva e negativa.
➡️A correlação positiva ocorre quando duas variáveis aumentam juntas.
Por exemplo, quanto mais um aluno estuda, maior é seu desempenho acadêmico.
➡️Já a correlação negativa acontecia quando uma variável aumenta enquanto a outra diminui.
Por exemplo, quanto mais tempo um aluno passa assistindo TV, menor é seu desempenho acadêmico.

A função de correlação mais usada é a de Pearson, que varia de -1 a 1. Um valor próximo de 1 indica uma forte correlação positiva, próximo de -1 indica uma forte correlação negativa,
e próximo de 0 indica que não há uma relação linear. Por exemplo, ao calcular a correlação entre horas de estudo e desempenho acadêmico, um coeficiente de 0.8 indica uma correlação positiva forte.
No nosso caso, calculamos a correlação entre idade e salário usando a função corr do pandas. Obtivemos um valor de 0.29, indicando uma correlação positiva, mas não muito forte.
📌Entender a correlação ajuda a evitar suposições.
Por exemplo, esperávamos que a idade fosse um fator importante para o aumento salarial, mas nossos dados mostram que, apesar de ter uma relação, não é tão forte.
Para identificar correlações em tabelas com muitas colunas de valores contínuos, utilizamos mapas de calor.
Um exemplo de mapa de calor permite visualizar rapidamente a correlação entre variáveis contínuas, com cores indicando a força da correlação.

'''

#Vamos fazer a correção - cor - que é uma função do pandas que fornece esses dados.
#Podemos observar que a idade não é um fator relevante para um maior salário, por exemplo.

correlacao_continua = dataframe['IDADE'].corr(dataframe['SALARIO'])
correlacao_continua
```

```python
#importando a biblioteca stats. Faz uma comparação entre a distribuição cruzada e esperada.

from scipy.stats import chi2_contingency
```

```python
'''
Também calculamos a correlação entre variáveis discretas usando o coeficiente de Cramér.
Criamos uma função cramer_coeficiente para calcular este coeficiente, utilizando a tabela cruzada gerada com crosstab
do pandas e a função chi2_contingency da biblioteca stats.
Aplicamos essa função para verificar a relação entre cor/raça/etnia e nível de ensino, obtendo um coeficiente de 0.044, indicando quase nenhuma relação.
Este resultado sugere que, na nossa amostra, etnias diferentes têm níveis de ensino semelhantes, possivelmente devido ao contexto da área de TI, onde a maioria já tem algum tipo de formação.
'''

def cramer_coeficiente(coluna1, coluna2):
  tabela_cruzada = np.array(pd.crosstab(coluna1, coluna2))
  chi2 = chi2_contingency(tabela_cruzada)[0] #retorna apenas o primeiro valor
  soma = np.sum(tabela_cruzada)
  mini = min(tabela_cruzada.shape)-1
  cramer = np.sqrt(chi2/(soma*mini))
  return cramer
```

```python
cramer_coeficiente(dataframe['COR/RACA/ETNIA'], dataframe['NIVEL DE ENSINO'])
# As colunas não tem quase nenhuma relação. Apesar do valor baixo, podemos dizer que as pessoas que estão na área de TI já possuem alguma formação
#De forma que a cor é algo que não tem muita relação com esse dado, por isso o valor foi tão baixo e a correlação não foi significativa.
```

```python
'''
Elementos Fundamentais para Gráficos Eficazes
➡️Tipo de Gráfico:
Escolha o tipo de gráfico que melhor represente seus dados e atenda ao seu objetivo.
1. Gráfico de Barras: Ideal para comparações, pode ser horizontal ou vertical, simples ou empilhado.
2. Gráfico de Pizza/Torta/Donut: Útil para comparações e visualização de proporções, com partes representando percentuais do total. Isso significa que a soma das partes nesses gráficos dão 100%
3. Gráfico de Linha: Exibe tendências ao longo do tempo, conectando pontos de dados. Ou seja, o eixo X é um valor de tempo: uma data, semana, dia mês. Mesmo quando não é temporal, utilizamos
com valores numéricos que tem alguma ordem, pois a linha é para continuidade.
4. Histograma: Analisa distribuições de dados, mostrando a frequência em intervalos.
5. Scatter Plot: Explora relações entre variáveis, cada ponto representando um dado.
6. Bubble Chart: Similar ao Scatter Plot, mas com bolhas de tamanhos variáveis para representar valores adicionais.

➡️Títulos e Rótulos:

Título Claro e Conciso: Comunique o objetivo da visualização de forma direta.
Rótulos nos Eixos: Identifiquem as variáveis representadas nos eixos X e Y.
Rótulos em Outros Elementos: Esclareçam a legenda de cores, formas ou outros elementos do gráfico.

➡️Eixos:

Eixos X e Y: Representam as variáveis dos dados de forma clara e precisa.
Rótulos nos Eixos: Indispensáveis para interpretar os dados.
Escala Apropriada: Permite a comparação correta entre os valores.

➡️Legenda:

Identifique Categorias ou Séries de Dados: Diferencie cada elemento do gráfico.
Clareza e Precisão: Utilize descrições fáceis de entender.
Posicionamento Adequado: Não obstrua a visualização dos dados.

➡️Cores:

Distinção de Categorias: Utilize cores para diferenciar elementos do gráfico.
Acessibilidade: Evite esquemas que dificultem a visualização para daltônicos.
Significado e Harmonia: Escolha cores que transmitam a mensagem desejada e sejam agradáveis aos olhos.

➡️Formas e Marcadores:

Identificação de Pontos de Dados: Utilize formas distintas para cada ponto de dados.
Padrões e Outliers: Facilite a identificação de padrões e outliers na visualização.
Clareza e Simplicidade: Evite formas complexas que dificultem a compreensão.

➡️Tamanho e Escala:

Tamanho Adequado: Permita a visualização clara dos elementos do gráfico.
Escala Proporcional: Represente os dados de forma precisa e comparável.
Evite distorções: Mantenha as proporções entre os elementos do gráfico.

➡️Interatividade (Opcional):

Explore os Dados de Forma Dinâmica: Permita que os usuários explorem os dados em detalhes.
Passe o Mouse sobre Pontos: Revele informações adicionais ao passar o mouse sobre cada ponto de dados.
Filtros e Seleções: Permita que os usuários filtrem e selecionem dados de acordo com suas necessidades.
'''
```

```python
'''
Nesta aula, exploramos o campo do Machine Learning (Aprendizado de Máquina), uma subárea da inteligência artificial que capacita os computadores a aprenderem padrões a partir de dados, sem serem explicitamente programados.
Discutimos diferentes tipos de algoritmos de aprendizado de máquina, como os supervisionados, não supervisionados, semi-supervisionados e por reforço🖥️
➡️Supervisionado: Um exemplo cotidiano é um algoritmo de recomendação de filmes, que aprende com dados de usuários (classificados como "gostei" ou "não gostei") para sugerir filmes similares.
➡️Não supervisionado: Um exemplo seria a segmentação de clientes em grupos com base em seus padrões de compra, sem rótulos prévios.
➡️Semi-supervisionado: Uma aplicação é a classificação de e-mails como spam ou não spam, onde apenas alguns e-mails são rotulados inicialmente, mas o algoritmo pode aprender com os padrões presentes nos dados.
➡️Por reforço: Pode ser exemplificado por um algoritmo que aprende a jogar xadrez, recebendo feedback positivo ou negativo com base nas jogadas.

Além disso, destacamos a diferença entre regressão e classificação:
➡️Regressão: Prevê valores contínuos. Por exemplo, prever o preço de uma casa com base em suas características, como tamanho, localização, número de quartos, etc.
➡️Classificação: Prevê classes ou categorias discretas. Um exemplo seria classificar e-mails como spam ou não spam, com base em seus atributos.
💭Propusemos um modelo de regressão simples neste módulo, e aprendemos que a regressão é um método estatístico usado para modelar a relação entre uma ou mais variáveis independentes e uma variável dependente.
Um exemplo cotidiano disso seria prever a quantidade de vendas de um produto com base no seu preço e nas despesas com publicidade.
💭Finalizamos a aula observando o fluxo de modelagem do modelo de regressão, entendendo que já fizemos a parte da análise dos dados em outros módulos, mas que estamos trabalhando agora
na parte de pré-processamento dos dados, antes de seguir para o treinamento do modelo.

No modelo de regressão temos um fluxo a ser seguido:
1. Análise Exploratória
2. Feature Engineering
3. Modelagem
4. Avaliação

Os dois primeiro são chamados de pré-processamento de dados
'''
```

###**7.2 - Preparação dos dados (parte 1)**

```python
'''
O modelo de regressão que faremos nesse módulo será apenas com os dados das pessoas que estão empregadas do tipo CLT.

➡️Então filtramos da coluna de “QUAL SUA SITUAÇÃO ATUAL DE TRABALHO?” todas as pessoas que marcaram a opção “Empregado (CLT)”.
➡️Depois disso criamos uma coluna chamada “NÃO_BRANCA”, aonde, se na coluna de “COR/RACA/ETNIA” for “branca” essa coluna recebe 0, e caso contrário recebe 1.
➡️Vamos para a coluna de “QUANTO TEMPO DE EXPERIÊNCIA NA ÁREA DE DADOS VOCÊ TEM?” e usamos a função extract para pegar o primeiro dígito
que aparece em cada opção dessa coluna. Aplicamos o mesmo para a coluna de “NÚMERO DE FUNCIONÁRIOS”.
'''
```

```python
#Para esse módulo usaremos um novo arquivo e vamos iniciar importando a biblioteca pandas

import pandas as pd
```

```python
#Importando os novos dados
dados = pd.read_excel('/content/drive/MyDrive/programaria - análise de dados/analise_dados_mod7_(1).xlsx')
```

```python
#Para visualizar o que tem nos dados que foram importados.
dados.head()
```

```python
#Visualizando o nome das colunas que iremos trabalhar

dados.columns
```

```python
dados['QUAL SUA SITUAÇÃO ATUAL DE TRABALHO?'].value_counts()
```

```python
#Nosso modelo vai se basear apenas em empregados CLTS, porque os outros possuem muitos parâmetros que podem dificultar a análise do modelo. Vamos fazer um filtro

dados = dados[dados['QUAL SUA SITUAÇÃO ATUAL DE TRABALHO?'] == 'Empregado (CLT)']
```

```python
#Mostrando apenas os empregados que são CLT
dados['QUAL SUA SITUAÇÃO ATUAL DE TRABALHO?'].value_counts()
```

```python
#Visualizando a coluna de cor, raça e etnia.

dados['COR/RACA/ETNIA'].value_counts()
```

```python
#Vamos retirar os valores que são muito pequenos, porque isso pode acabar interferindo no geral, uma amostra de 4 indígenas é muito pequena perto da quantidade de representantes do grupo.
#Para retirar esses valores:

lista_retirar = ['Prefiro não informar', 'Outra', 'Indígena']
```

```python
#Iremos fazer um filtro para retirar os itens anteriores e selecionar alguns parâmetros que serão utilizados no nosso modelo

dados = dados[~ dados ['COR/RACA/ETNIA'].isin(lista_retirar)]
```

```python
dados['NAO_BRANCA'] = dados['COR/RACA/ETNIA'].apply(lambda x: 1 if x != 'Branca' else 0)
```

```python
dados['QUANTO TEMPO DE EXPERIÊNCIA NA ÁREA DE DADOS VOCÊ TEM?'].value_counts()
```

```python
#Da tabela acima, vamos pegar o primeiro valor de cada item. Criamos uma coluna chamada tempo de experiência usando a função extract para extrair os parâmetros

dados['TEMPO_EXPERIENCIA'] = dados['QUANTO TEMPO DE EXPERIÊNCIA NA ÁREA DE DADOS VOCÊ TEM?'].str.extract(r'(\d+)')
```

```python
#Visualizando a coluna que acabamos de criar

dados['TEMPO_EXPERIENCIA']
```

```python
dados['TEMPO_EXPERIENCIA'].value_counts()
```

```python
#Verificando se tem números nulos, como tem, podemos substituir por zero.

dados['TEMPO_EXPERIENCIA'].value_counts(dropna = False)
```

```python
#Substituindo os valores nulos por zero

dados['TEMPO_EXPERIENCIA'] = dados['TEMPO_EXPERIENCIA'].fillna(0)
```

```python
#Verificando se foram excluidos os valores nulos
dados['TEMPO_EXPERIENCIA'].value_counts(dropna = False)
```

```python
#Coluna número de funcionários

dados['NUMERO DE FUNCIONARIOS'].value_counts()
```

```python
#Retirando o ponto
dados['NUMERO DE FUNCIONARIOS'] = dados['NUMERO DE FUNCIONARIOS'].str.replace('.', '')
```

```python
#Verificando se foi retirado o ponto mesmo

dados['NUMERO DE FUNCIONARIOS'].value_counts()
```

```python
#Finalmente podemos usar o extract

dados['NUMERO DE FUNCIONARIOS'] = dados['NUMERO DE FUNCIONARIOS'].str.extract(r'(\d+)')
```

```python
#Verificando se ocorreu a substituição

dados['NUMERO DE FUNCIONARIOS'].value_counts()
```

```python
#Para retirar, caso tenha, valores nulos

dados['NUMERO DE FUNCIONARIOS'].value_counts(dropna = False)
```

###**7.2 - Preparação dos dados (parte 3)**

```python
'''
➡️Após fazer as alterações nas colunas individuais, selecionamos as colunas que serão os argumentos do nosso modelo de regressão: 'IDADE', 'GENERO', 'NAO_BRANCA','TEMPO_EXPERIENCIA', 'SETOR',
'REGIAO ONDE MORA','NIVEL DE ENSINO',  'NUMERO DE FUNCIONARIOS', 'NOVO_NIVEL','SALARIO' e 'INSATISFACAO'.
➡️Aplicamos o get_dummies para as colunas de 'GENERO', 'REGIAO ONDE MORA' e 'NOVO_NIVEL'. Lembrando que usamos o get_dummies aqui pois as categorias não têm uma ordem crescente como em nivel de ensino.
➡️Antes de treinar o modelo em si, nós dividimos os dados em dois conjuntos, um de teste (para avaliação do modelo) e um de treinamento (para o treinamento do modelo).
➡️Separamos os atributos do nosso objetivo (target = salário), e utilizamos uma função chamada train_test_split para fazer a divisão dos dados.
➡️Depois que a divisão foi realizada, nós usamos a função StandardScaler para padronizar as características, ou seja, normalizar os nossos dados.
➡️Após finalizado a parte de pré processamento dos dados, podemos seguir para o treinamento de modelagem.

'''
```

```python
#Iniciamos agora fazendo a seleção das colunas que iremos utilizar no nosso modelo

dados = dados[['IDADE', 'GENERO', 'NAO_BRANCA', 'TEMPO_EXPERIENCIA',
              'SETOR', 'REGIAO ONDE MORA', 'NIVEL DE ENSINO',
              'NUMERO DE FUNCIONARIOS', 'NOVO_NIVEL', 'SALARIO', 'INSATISFACAO']]
```

```python
dados.columns
```

```python
#Get_dumies - em situações que não tem como atribuir valores. O Drop_first elimina a primeira coluna

pd.get_dummies(dados, columns = ['GENERO', 'REGIAO ONDE MORA', 'NOVO_NIVEL', 'SETOR'], drop_first = True)
```

```python
dados = pd.get_dummies(dados, columns = ['GENERO', 'REGIAO ONDE MORA', 'NOVO_NIVEL', 'SETOR'], drop_first = True)
```

```python
#Vamos separar o nosso conjunto de dados em dois. Uma parte vai ser utilizada como conjunto de treinamento para o modelo aprender e o outro de teste, para ver o que modelo ainda não viu.

x = dados.drop('SALARIO', axis=1)  # Exclude 'SALARIO' if it's your target variable
y = dados['SALARIO']
```

```python
#Importando uma biblioteca de treino e teste

from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42) #20% dos dados para teste e os outros 80% para treinamento
```

```python
#Importando outra biblioteca do sklearn - normaliza os dados

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

###**7.3 - Introdução a regressão linear (parte 1)**

```python
'''
➡️Nós importamos a função LinearRegression da biblioteca sklearn, criamos um objeto chamado model apartir dessa função. E então enviando os parâmetros de
treino na função fit (assim: model.fit(X_train_scaled, y_train)) nós conseguimos treinar o modelo.
➡️Usando a função predict e enviando como parâmetro o conjunto de teste (X_test_scaled) a gente conseguiu fazer as predições desse conjunto.
➡️E então partimos para uma parte muito importante de qualquer modelo de machine learning que é a avaliação do modelo, para saber se está bom ou ruim (ou mais ou menos, ou médio, ou excelente, ou … enfim,
vocês entenderam, certo?😂) Focamos em três métricas principais para avaliação de modelos de regressão:
MSE (Mean Squared Error), MAE (Mean Absolute Error) e R² (R-squared). Com os resultados nós concluímos que para um modelo real de produção nosso modelo não seria aceitável.


'''
```

```python
#Nessa etapa vamos iniciar o treinamento do nosso modelo
#Vamos importar as bibliotecas necessárias

from sklearn.linear_model import LinearRegression
```

```python
#Realizando o treniamento
model = LinearRegression()
```

```python
model.fit(x_train_scaled, y_train) #Treinamento do modelo - resultado é o modelo treinado
```

```python
y_pred = model.predict(x_test_scaled) #Fazendo as predições
```

```python
model.predict(x_test_scaled) #Os valores obtidos são os valores de salário que o modelo predizeu para cada item
```

```python
#Avaliação do modelo usando as métricas MSE
#Importando as biliotecas necessárias

from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
#Vamos calcular MSE

mse = mean_squared_error(y_test, y_pred)
```

```python
mse #Valor muito alto, dizendo que o modelo não está muito bom. Porém a métrica é diferente do nosso salário, não sendo muito bom para esse tipo de dado
```

```python
#Vamos analisar com base em outra métrica - Erro médio absoluto

mae = mean_absolute_error(y_test, y_pred)
```

```python
mae
```

```python
#Olhando o R²

r2 = r2_score(y_test, y_pred)
```

```python
r2
```

```python
import pandas as pd


pd.DataFrame(model.coef_, columns = ['Coeficientes'], index = nomes_atributos) # Change 'dataframe' to 'DataFrame'
```

```python
#Coeficientes do modelo que treinamos, estamos dando um nome para a coluna e, por fim, as linhas serão referenciadas pelos atributos. Estamos armazenando tudo na variável coefs.
coefs = pd.DataFrame(model.coef_, columns = ['Coeficientes'], index = nomes_atributos)
```

```python
#Vamos ordenar essa coluna para verificar o que está influenciando mais. E em seguida o resultado será salvo no coefs

coefs = coefs.sort_values(by = 'Coeficientes', ascending = False)
```

```python
#Vamos plotar um Gráfico para verificar a influência de forma mais fácil
#Todos os valores antes do zero tem um peso negativo no salário, ajudando a diminuir os valores.

coefs.plot.barh(figsize = (8, 6))
plt.axvline(x = 0, color = 'red')
plt.title('Coeficientes do Modelo')
plt.show()
```
